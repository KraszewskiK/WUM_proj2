### Widać, że zdecydowana większość wartości to zera.

```python
l = []
for i in df.columns[1:]:
    l.append(sum(df[i] == 0))
print("Średnia zawartość zer w kolumnie (w %): ", np.mean(l)/590)
print("Minimalna zawartość zer w kolumnie (w %): ", min(l)/590)
print("Średnia zawartość zer w kolumnie (w %): ", max(l)/590)
```

```python
def count_zeroes(x):
    y = []
    for x_i in x:
        c = 0
        for i in df.columns:
            if sum(df[i] != 0) > x_i:
                c += 1
        y.append(c)
    return y

x = range(1,50)
plt.figure()
plt.plot(x, count_zeroes(x))
plt.title("Liczba kolumn względem liczby wartości niezerowych")
plt.show()
```

## Redukcja danych

```python
l = []
for i in df.columns:
    if sum(df[i] != 0) > 118:
        l.append(i)
print("Liczba kolumn, w których co najmniej 20% wartości jest niezerowych: ", len(l))
```

```python
df_reduced = df[l]
df_reduced
```

```python
df_reduced.describe()
```

```python
df_reduced.hist(bins = 30, figsize = (18,12))
plt.show()
```
